```python
import torch
from cp_dataset import VITONDataset
```

```python
dataset = VITONDataset(
    root_dir="./dataset",
    pairs_file="train_pairs.txt",
    mode="train",
    img_size=(512, 384)
)
```

```python
len(dataset)
```

```python
if len(dataset) > 0:
        # İlk örneği al
        sample = dataset[0]
        
        print("\nYüklenen veriler:")
        for key, value in sample.items():
            if isinstance(value, torch.Tensor):
                print(f"  {key}: shape={value.shape}, dtype={value.dtype}")
            else:
                print(f"  {key}: {value}")
        
        # DataLoader örneği
        from torch.utils.data import DataLoader
        
        dataloader = DataLoader(
            dataset,
            batch_size=4,
            shuffle=True,
            num_workers=2,
            pin_memory=True
        )
        
        # Bir batch al
        batch = next(iter(dataloader))
        print(f"\nBatch boyutları:")
        for key, value in batch.items():
            if isinstance(value, torch.Tensor):
                print(f"  {key}: {value.shape}")
```
